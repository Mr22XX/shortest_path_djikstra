```python
import openrouteservice
import folium

# === 1. API KEY ORS ===
client = openrouteservice.Client(key='5b3ce3597851110001cf6248aaa63c5678d84f509bc13522adb804d3')

# === 2. Koordinat lokasi ===
locations = {
    "Rektorat": [102.27231460986346, -3.7590495172423495],
    "Masjid Baitul Hikmah": [102.27600666694858, -3.758945132312725],
    "Perpustakaan" : [102.27485462111163, -3.756806076798016],
    "Gerbang Masuk Belakang" : [ 102.27521569504947, -3.759614988399855]
}

# === 3. Ambil koordinat untuk ORS ===
start_input = input("Input Lokasi Awal : ")
end_input = input("input Lokasi Akhir : ")
start_coord = locations[start_input]
end_coord = locations[end_input]
route_coords = [start_coord, end_coord]

# === 4. Request rute dengan alternatif ===
routes = client.directions(
    coordinates=route_coords,
    profile='driving-car',
    format='geojson',
    alternative_routes={"share_factor": 0.6, "target_count": 3}
)

# === 5. Buat peta dengan titik tengah kampus UNIB ===
m = folium.Map(location=[-3.7593271927243426, 102.27239872520606], zoom_start=17)

# === 6. Tambah marker lokasi asal dan tujuan ===
for name, coord in locations.items():
    folium.Marker(
        location=[coord[1], coord[0]],
        popup=name,
        icon=folium.Icon(color='blue' if name == 'Rektorat' else 'green')
    ).add_to(m)

# === 7. Warna untuk rute ===
# === Red untuk Shortest path & Warna lain untuk alternative path ===
colors = ['red', 'blue', 'green']

# === 8. Tampilkan semua rute alternatif di peta dan cetak info ===
print("=== PERBANDINGAN JARAK DAN WAKTU SETIAP RUTE ===\n")

for i, feature in enumerate(routes['features']):
    coords = feature['geometry']
    properties = feature['properties']['summary']
    distance_km = properties['distance'] / 1000
    duration_min = properties['duration'] / 60

    # Tampilkan ke map
    folium.GeoJson(
        feature,
        name=f'Rute {i+1}',
        tooltip=f'Rute {i+1}: {distance_km:.2f} km, {duration_min:.2f} menit',
        style_function=lambda x, col=colors[i % len(colors)]: {
            'color': col, 'weight': 5, 'opacity': 0.8
        }
    ).add_to(m)

    # Tampilkan info ke terminal
    print(f"Rute {i+1}:")
    print(f"  - Jarak  : {distance_km:.2f} km")
    print(f"  - Waktu  : {duration_min:.2f} menit")
    print()

# === 9. Tampilkan peta ===
m
```
